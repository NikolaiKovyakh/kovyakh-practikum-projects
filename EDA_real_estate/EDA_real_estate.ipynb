# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

Проект будет состоять из нескольких этапов:
1. Изучение общей информации о данных.
2. Проведение предобработки данных.
3. Изменение таблицы, путём добавления в неё новых столбцов.
4. Исследовательский анализ данных.

В проекте необходимо изучить:
1. Как быстро продавались квартиры. Этот параметр показывает, сколько дней было размещено каждое объявление.
2. Какие факторы больше всего влияют на общую (полную) стоимость объекта?
3. Выделить 10 самых населённых пунктов и найти там пункты с самой высокой и низкой стоимостями квадратного метра. 
4. Описать, как стоимость объектов в Санкт-Петербурге зависит от расстояния до центра города.



**Описание данных**
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

В столбце `total_area` не наблюдается каких-то аномалий и пропущенных значений.\

```python
living_percent = df['living_area'] / df['total_area'] * 100
living_percent.hist()
plt.xlabel('percentage of living area of the total area');
plt.ylabel('count');
```

Чтобы заменить пропущенные значения в картографических данных, лучше понять, какие данные к какому району относятся.
То есть нельзя заменять пропуски в числовых картографических данных на медианные по всему датасету. 
Лучше заменить пропуски на медианные значения по регионам. 
Так как скорее всего количество озёр или парков рядом с одной квартирой и другой из одного и того же района
должно совпадать.\
Но для того, чтобы сделать такую замену, сначала необходимо избавиться от дубликатов в названиях населённых пунктов.\
Выведем количество населённых пунктов с дубликатами

```python
df['locality_name'].nunique()
```

```python
print(*sorted(df['locality_name'].unique()), sep='\n')
```

В данных заметно, что некоторые одинаковые слова пишутся то через букву "ё", то через "е".\
Сделаем замену

```python
df['locality_name'] = df['locality_name'].str.replace('ё','е')
```

```python
df['locality_name'].nunique()
```

Количество уникальных значений сократилось с 365 до 331!

Также заметим, что лучше заменить разные виды посёлков на просто "поселок".\
Заведём список со всеми неявными дубликатами для названия "поселок". 
Это делаем для того, чтобы потом заменить посёлки разных типов на просто "поселок".

```python
copies = ['поселок городского типа', 'городской поселок', 'коттеджный поселок', 
          'поселок станции', 'поселок при железнодорожной станции']
```

```python
for c in copies:
    df['locality_name'] = df['locality_name'].str.replace(c, 'поселок')
df['locality_name'].nunique()
```

Количество уникальных значений ещё немного сократилось.

Оставим все пропуски в картографических данных без изменений.

Сделаем финальную проверку, сколько и где осталось пропущенных значений

```python
df.isna().sum()
```

Причины возникновения пропусков в любой колонке могут быть как ошибкой при выгрузке базы данных, так и неверном заполнении базы данных, а ещё из-за человеческих ошибок при формировании базы данных. 
- Изначально данные о количестве балконов в объекте могут отсутствовать, потому что объект не имеет балкона.
- Высота потолков могла быть не указана, потому что не была измерена изначально. 
- Жилая площадь могла быть не указана, потому что при заполнении не удалось выяснить, какая часть площади - жилая, а какая - нет, или потому что при заполнении таблицы могли посчитать, что общая площадь и жилая - это одно и то же - в итоге эти поля остались незаполненными.
- Картографические данные могут быть незаполнены, потому что их было тяжело собрать и измерить. Для очень многих населённых пунктов, возможно, не имеет значения, какое расстояние до аэропорта или центра, потому что расстояния достаточно велики, чтобы их не стали вносить или потому что они не имеют особого значения. 

Причин может быть много.

#### Изменение типов данных

```python
df.info()
```

Преобразуем значения количества числа этажей и балконов (а также количество фото, комнат и этажей - ради экономия памяти) к целам числам int8. Всем этим переменным такого формата будет достаточно.

```python
df[['floors_total', 'balcony', 'total_images', 'rooms', 'floor']] = df[['floors_total', 'balcony', 
                                                               'total_images', 'rooms', 'floor']].astype('int8')
```

Число водоёмов, парков и дней до снятия с публикации тоже неплохо бы привести к целочисленному типу (int32), но это не получится сделать, потому что 
эти поля имеют пропуски 'NaN', NaN - невозможно преобразовать к int32.
Но их можно преобразовать к типу Int64! И тогда пропущенные значения станут записываться как '< NA >'

```python
df[['parks_around3000', 'ponds_around3000']] = df[['parks_around3000', 'ponds_around3000']].astype('Int64')
```

```python
try:
    df['days_exposition'] = np.floor(pd.to_numeric(df['days_exposition'], errors='coerce')).astype('Int64')
except:
    print('Данные не получается преобразовать!')
```

Заметим также, что колонка first_day_exposition - это тип "object", хотя логичнее её сделать датой. 
Причём точное время в часах и минутах не так важно, оно почти всегда 00:00 - и не несёт в себе полезной информации.  
Преобразуем first_day_exposition в дату.

```python
df['first_day_exposition'] = pd.to_datetime(
    df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Для экономии памяти также переведём данные с вещественными значениями из типа float64 в float32, а данные с целыми значениями в int32.

```python
for col in df.columns:
    if pd.api.types.is_float_dtype(df[col]):
        df[col] = df[col].astype('float32')
```

```python
def floor_type(row):
    if row['floor'] == row['floors_total']:
        return 'последний'
    elif row['floor'] == 1:
        return 'первый'
    return 'другой'
```
