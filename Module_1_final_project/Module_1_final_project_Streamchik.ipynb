### Изменение типов данных и обработка пропусков

Приведём названия столбцов к нижнему регистру

```python
df.columns = df.columns.str.lower()
display(df.columns)
```

Заметим, что типы не для всех переменных выбраны подходящие типы данных. Год выпуска (year_of_release) - вещественный тип (float64), когда лучше сделать его целым (int). Оценка пользователей (user_score) - строковый тип (object), лучше сделать его вещественным (float32).

```python
display(df['year_of_release'].unique())
display(df['user_score'].unique())
display(df['critic_score'].unique())
```

В столбцах year_of_release (как и в critic_score) есть пропущенные значения, поэтому не получится сразу преобразовать год к целому типу.

В столбце user_score есть не только пропущенные значения nan, но и особый вид неявных пропусков 'tbd', который, вероятно, означает, что у игры рейтинг должен появиться позже. To be decided или to be determined означает, что информация об объекте неизвестна и может быть дополнена в будущем, но в нашем случае мы должны обработать это тоже как пропуски.

```python
print('Всего пропусков типа "tbd" в рейтинге пользователей:', len(df[df['user_score'] == 'tbd']))
print('Обычных пропусков типа nan:', df['user_score'].isna().sum())
print('Всего пропусков:', len(df[df['user_score'] == 'tbd']) + df['user_score'].isna().sum())
```

Девять тысяч пропусков - это огромная цифра, больше половины данных.
Заменим пропуски 'tbd' на стандартное None в python.

```python
df.loc[df['user_score'] == "tbd", 'user_score'] = None
```

Теперь приведём столбец user_score к типу float32.

```python
df['user_score'] = df['user_score'].astype('float32')
```

Чтобы перевести тип данных столбца year_of_release к целочисленному понадобится сначала обработать пропуски

Обработка пропусков в year_of_release

Например, в серии FIFA Soccer год после названия не означает год выпуска, так как игра с номером n обычно имеет год выпуска n-1. Например, FIFA Soccer 2004 выходила в 2003 году и так далее.

Проверим, сколько есть в таблице игр, у которых невозможно посчитать среднее значение года выпуска

```python
na_in_years = df.pivot_table(index='name', 
                             values='year_of_release', 
                             aggfunc='mean', dropna=False).reset_index()['year_of_release'].isna().sum()
print(f"для {na_in_years} игр нельзя посчитать среднее или медиану года")
```

Заменим пропуски в годах медианными значениями для каждой игры. Это получится сделать не для всех игр.

```python
# Чтобы не появлялись уведомления о предупреждениях (warnings) отключим их на время восполнения пропусков
warnings.filterwarnings("ignore")
df['year_of_release'] = df[df['name'].notna()].groupby("name")['year_of_release']\
    .transform(lambda x: x.fillna(round(x.median(), 0)))
warnings.filterwarnings("default")
```

```python
print(f'Для {sum(df["year_of_release"].isna())} строчек в таблице не удалось заменить пропуски медианными значением') 
```

Изменим тип данных года на целочисленный, предварительно избавившись от пропусков. Их всего 148, не будем их обрабатывать.

```python
df = df[df['year_of_release'].notna()]
df['year_of_release'] = df['year_of_release'].astype('int32')
```

```python
print(df['critic_score'].unique())
print(df['user_score'].unique())
```

Аномалий не наблюдаем.

Обработам возрастные рейтинги. Если они не известны для игры на конкретной платформе, то скопируем для этой же игры первый попавшийся рейтинг с другой платформы

```python
# Создадим таблицу для игр с известными рейтингами, 
# выбрав первый попавшийся рейтинг для конкретной игры методом first()
# но назовём колонку с рейтингом по другому, 
# чтобы потом её можно было добавить в исходный датафрейм
ratings = df[['name', 'rating']].dropna()\
    .pivot_table(index='name', values='rating', aggfunc='first')\
    .reset_index().rename(columns={'rating': 'known_rating'})
ratings.head()
```

```python
# добавим новый столбец к исходной таблице методом присоединений таблиц merge() 
df = df.merge(ratings, on="name", how="left")
df.head()
```

```python
# скорпируем из столбца known_rating не NaN-рейтинги в столбец rating только для тех записей, которые имеют рейтинг NaN
df.loc[(df['rating'].isna() & (df['known_rating'].notna())), 'rating'] = df['known_rating']
# удаляем лишнюю переменную known_rating
df.drop(columns=['known_rating'], inplace=True)
```

```python
df.isna().sum()
```

Пропусков в рейтингах стало почти на 400 записей меньше, но их всё ещё много.

Заменим оставшиеся пропуски в возрастных рейтингах записью 'unknown'

```python
df['rating'] = df['rating'].fillna('unknown')
df['rating'].unique()
```

Узнаем, какие уникальные рейтинги есть в таблице и какое их количество каждого типа

```python
df['rating'].value_counts()
```

Чтобы лучше разобраться в отличиях, сначала поясним сокращённые обозначения:
- E - контент для всех возрастов
- E10+ - контент для лиц от 10 лет. Может содержать изображение насилия в анимационной, фентезийной форме или незначительное его количество. А также незначительное количество ругательств
- T - контент для лиц от 13 лет. Может содержать насилие, наводящие темы, грубый юмор, минимальное количество крови, симулировать азартные игры и нечастые использования ругательств.
- M - контент для лиц от 17 лет. Может содержать частые насилие, кровь, жестокость, сексуальный контент и ругательства.
- unknown - это рейтинг, который неизвестен. Это категория, которую мы заполнили самостоятельно вместо пропусков.
- EC - формат для малых детей от 3-х лет. Контент не содержит неприемлимого материала. 
- K-A - это рейтинг для Kids-Adults (Детей и взрослых) - этот рейтинг существовал до 1998 года, а после ESRB советовала использовать рейтинг "Е"
- RP - Rating Pending - рейтинг ожидается. Этот символ появляется только в рекламе перед выпуском игры.
- AO - Adults Only - игры только для взрослых, от 18+ лет.

Заметим, что рейтинг EC очень схож с рейтингом Е - на него и заменим эти редкие значения. Рейтинг K-A даже сама организация ESRB советует заменить на E. RP - это рейтинг ожидается, но он нам неизвестен, заметим его на unknown. AO - такая игра единственная, можно её удалить.

```python
df.loc[(df['rating'] == 'K-A') | (df['rating'] == 'EC'), 'rating'] = 'E'
df = df[df['rating'] != 'AO']
df.loc[df['rating'] == 'RP', 'rating'] = 'unknown'
df['rating'].value_counts()
```

```python
df.isna().sum()
```

## Исследовательский анализ данных

Проверим сколько игр выпускалось за разные годы. Можно предположить, что в таблице будет мало игр, например, 1980-го года

```python
df.pivot_table(index='year_of_release', values='name', aggfunc='count')\
    .reset_index().rename(columns={'name': 'numbers'})\
    .plot(kind='bar', x='year_of_release', y='numbers', figsize=(7, 5),  legend=None);
plt.xlabel('Year', fontsize=15);
plt.ylabel('Count games', fontsize=10);
plt.title('Count games by year', fontsize=10);
plt.tick_params(which='major', labelsize=10);
```

По графику видно, что в базе очень мало игр 1980-1994 годов. Попробуем посчитать их процент.

```python
print(f"Игр до 1995 года: {len(df[df['year_of_release'] <= 1994]) / len(df):.2%}")
```

```python
# Слегка актуализируем данные. Не будем брать года раньше 1995-го
actual_df = df[df['year_of_release'] > 1994]
```

Посмотрим, как менялись продажи по 10 самым большим платформам (у которых самые большие суммарные продажи).

```python
top_sales_platforms = actual_df.pivot_table(index='platform', values='total_sales', aggfunc='sum').reset_index()\
    .sort_values(by='total_sales', ascending=False).reset_index(drop=True)[:10]

top_platforms = top_sales_platforms['platform']

data = actual_df[actual_df['platform'] == 'PS']\
    .pivot_table(index='year_of_release', values='total_sales', aggfunc='sum').reset_index()\
    .rename(columns={'total_sales': 'PS sales'})

for model in top_platforms:
    data = data.merge(
        actual_df[actual_df['platform'] == model]\
        .pivot_table(index='year_of_release', values='total_sales', aggfunc='sum').reset_index()\
        .rename(columns={'total_sales': f'{model} sales'}), how='outer'
        )

data = data.set_index('year_of_release')

data.plot(style='o-', grid=True, figsize=(9, 7));
plt.xticks(rotation = 60)
plt.xticks(data.index)
plt.xlabel('Year', fontsize=16);
plt.ylabel('Game sales, millions', fontsize=16);
plt.title('Game sales by platforms', fontsize=16);
plt.tick_params(which='major', labelsize=12);
```

На таком графике видно, что в среднем жизненный цикл одной платформы составляет около семи лет. Например, PS (Playstation) в наших данных появляется в 1995-м году, продажи растут до 1998-го года, далее идёт небольшой спад до 2000-го года, и уже в 2001-м году продажи игр на этой платформе падают до уровня запуска платформы. Такая же ситуация и с PS2 - старт продаж в 2000, пик в 2002-2004, падение продаж в 2008. Та же ситуация и с другими консолями. На этом графике всего 7 платформ, но по нему видно, что спад продаж игр на одну платформу всегда сопровождается ростом продаж игр на других, более новых платформах.

Таким образом, попробуем определить данные за актуальный период, то есть изучить последние 8 лет продаж. 8 лет - за такой период должны стать видны падения продаж платформ уходящего поколения и рост продаж новых.
```python
# Ещё чуть актуализируем данные
actual_df = actual_df[actual_df['year_of_release'] >= 2009]
# Выведем все платформы, о которых есть данные за 2009-2016 года
actual_df['platform'].unique()
```

```python
# Создадим топ 10 платформ за последние 8 лет
top_sales_platform = actual_df.pivot_table(index='platform', values='total_sales', aggfunc='sum').reset_index()\
    .sort_values(by='total_sales', ascending=False).reset_index(drop=True)[:10]
top_sales_platform
```

## Проверка гипотез

### Отличия в средних пользовательских рейтингах на платформах Xbox One и PC

Чтобы сравнения были осмысленными, будем сравнивать только игры за 2013-2016 год. Данные для XOne доступны только для этих годов. И в таком случае количество игр из одной категории и другой не должно сильно отличаться.

```python
print("Игр на XOne:", len(actual_df.loc[actual_df['platform'] == 'XOne', 'user_score']))
print("Игр на PC:", len(actual_df.loc[actual_df['platform'] == 'PC', 'user_score']))
```

```python
mean_PC = actual_df.loc[actual_df['platform'] == 'PC', 'user_score'].mean()
print('Средний пользовательский рейнтиг на PC:', mean_PC)
```

```python
mean_xone = actual_df.loc[actual_df['platform'] == 'XOne', 'user_score'].mean()
print('Средний пользовательский рейнтиг на Xbox One:', mean_xone)
```

Согласно Центральной предельной теореме выборочное среднее распределено нормально вокруг истинного среднего генеральной совокупности. Выборки не равны, дисперсии не равны. 

Наиболее вероятно, что мы не знаем истинных средних двух генеральных совокупностей и хотим сравнить, равны ли они. Воспользуемся t-тестом. Для этого используем метод ttest_ind() для проверки гипотезы о равенстве средних двух генеральных совокупностей по взятым из них выборкам, тогд будем считать, что:
- нулевая гипотеза **H0**: "Средний рейтинг игр на Xbox One = Средний рейтинг игр на PC"
- альтернативная **H1**: "Средний рейтинг игр на Xbox One != Средний рейтинг игр на PC"
<br><br>Гипотеза двухсторонняя.

```python
import scipy.stats as st
```

```python
# Запишем в отдельную переменную все оценки игр для Xbox One
xone_scores = list(actual_df.loc[(actual_df['platform'] == 'XOne') & (actual_df['user_score'].notna()), 'user_score'])
pc_scores = list(actual_df.loc[(actual_df['platform'] == 'PC') & (actual_df['user_score'].notna()), 'user_score'])
```

```python
# Задаидим уровень значимости 0.05
alpha = .05
# Уровень значимости численно определит, 
# где проходит граница вероятности попасть слишком далеко от центра распределения.
results = st.ttest_ind(xone_scores, pc_scores, equal_var=False)
print('p-значение:', results.pvalue)

# Если p-value будет будет очень мало, то это будет означать, что вероятность получить наблюдаемый 
# или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна - очень мала. 
# Вероятность мала - значит, этоm результат не совпадение не совпадение и нулевую 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Рейтинги игр на Xbox One не равны рейтингам игр на ПК")
else:
    
    print("Вероятность получить такие отличия в рейтинге достаточно высока. Не получилось отвергнуть нулевую гипотезу")
```

Принять нулевую гипотезу мы не можем, но и отвергнуть тоже. **Не удалось найти значимых различий в рейтингах игр на Xbox One и PC**

### Отличия средних пользовательских рейтингов жанров Action и Sports

```python
print("Игр в жанре Action:", len(actual_df.loc[actual_df['genre'] == 'Action', 'user_score']))
print("Игр в жанре Sports:", len(actual_df.loc[actual_df['genre'] == 'Sports', 'user_score']))
```

Игр в жанре Экшен гораздо больше, чем игр в жанре спорт

```python
mean_action = actual_df.loc[actual_df['genre'] == 'Action', 'user_score'].mean()
mean_action
```

```python
mean_sports = actual_df.loc[actual_df['genre'] == 'Sports', 'user_score'].mean()
mean_sports
```

```python
print(f'Средний пользовательский балл спортивных игр на {mean_action - mean_sports:.1f} баллов ниже, чем у экшенов.')
```

Тут **различия настолько значительны**, что можно было бы и без проверок сказать, что **игры в жанре спорт имеют рейтинги намного хуже, чем экшены**. Но всё-таки проверерим t-тестом.

```python
action_scores = list(actual_df.loc[(actual_df['user_score'].notna()) & (actual_df['genre'] == 'Action'), 'user_score'])
sports_scores = list(actual_df.loc[(actual_df['user_score'].notna()) & (actual_df['genre'] == 'Sports'), 'user_score'])
```

**Мы не знаем истинные средние** - и хотим сравнить, равны ли они. Для этого используем метод ttest_ind() для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Рейтинг экшен игр кажется намного выше, поэтому попробуем проверить одностороннюю гипотезу о том, что он именно выше, а не просто не равен рейтингу спортивных игр.

В этом случае будем считать, что: 
- нулевая гипотеза **H0**: "Средний пользовательский рейтинг игр в жанре экшен = Средний пользовательский рейтинг игр в жанре спорт"
- альтернативная **H1**:  "Средний пользовательский рейтинг игр в жанре экшен > Средний пользовательский рейтинг игр в жанре спорт"
<br><br>Гипотеза односторонняя.

Уровень значимости оставим тот же самый. Проверим гипотезу при условии, что дисперсии неравны.

```python
results_2 = st.ttest_ind(action_scores, sports_scores, equal_var=False, alternative='greater')

print('p-значение:', results_2.pvalue)
if results_2.pvalue < alpha:
    print(f"Отвергаем нулевую гипотезу. Вероятность получить такие отличия в рейтинге при условии, что верна H0 почти равна нулю \
(при округлении до десяти знаков p = {results_2.pvalue:.10f}). \
Рейтинги экшен игр больше рейтингов спортивных игр")
else:
    print("Вероятность получить такие отличия в рейтинге достаточно высока. Не получилось отвергнуть нулевую гипотезу") 
```

**Выводы по гипотезам**: 
1) В cредних пользовательских рейтингах платформ Xbox One и PC **не** удалось найти значимых различий.
<br>2) Средние пользовательские рейтинги жанров Action и Sports **не** одинаковые. Рейтинги экшен игр больше рейтингов спортивных игр
