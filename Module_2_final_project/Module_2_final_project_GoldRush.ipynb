### Проверка, что эффективность обогащения рассчитана правильно. 

Для этого рассчитаем по имеющимся столбцам самостоятельно показатель Recovery, а потом посчитаем MAE между собственноручно вычисленным Recovery для каждого этапа и известными соответствующими столбцами final.output.recovery и rougher.output.recovery

Рассчитаем MAE для сравнения эффективности **обогащения после всех этапов** из таблицы и подсчитанной вручную эффективности после всех этапов

```python
F = train['rougher.input.feed_au']
C = train['final.output.concentrate_au']
T = train['final.output.tail_au']
RECOV = 100 * C * (F - T) / (F * (C - T))
mean_absolute_error(train['final.output.recovery'], RECOV)
```

```python
train['final.output.recovery'].head(), RECOV.head()
```

Рассчитаем MAE для сравнения эффективности **обогащения после флотации** из таблицы и подсчитанной вручную эффективности после флотации.

```python
C = train['rougher.output.concentrate_au']
T = train['rougher.output.tail_au']
RECOV = 100 * C * (F - T) / (F * (C - T))
mean_absolute_error(train['rougher.output.recovery'], RECOV)
```

Средняя ошибка в двух случаях очень мала. Если вывести столбцы, то будет видно, что все показатели совпадают с точностью до шести знаков после запятой. Небольшая ошибка скорее всего просто связана со спецификой хранения дробных чисел в python.

## Модель

### Функция для вычисления итоговой sMAPE

sMAPE (*англ.* **Symmetric Mean Absolute Percentage Error**, «симметричное среднее абсолютное процентное отклонение»).
Метрика похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Симметричная - значит, одинаково учитывает масштаб и целевого признака, и предсказания.

```python
# Сначала напишем функцию для вычисления smape
def smape(y_true, y_pred):
    return (2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred))).mean() * 100
```

```python
# функция для вычисления финального sMape
def final_smape(rough_smape, final_smape):
    return 0.25 * rough_smape + 0.75 * final_smape
```

```python
sMAPE = make_scorer(smape, greater_is_better=False)
```

### Обучение разных моделей и оценка их качества кросс-валидацией. 

Нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата rougher.output.recovery;
- эффективность обогащения финального концентрата final.output.recovery.

Моделей будем рассматривать много - 10 для предсказания эффективности обогащения чернового концентрата (5 - для данных без коррелированных столбцов, 5 для данных с ними) и 10 для предсказания эффективности обогащения финального концентрата. Поэтому заведём пустой сводный датафрейм, в который будем заносить лучшую метрику sMAPE для каждой из моделей.

```python
all_models = pd.DataFrame([[0, 0, 0, 0]] * 5, columns=['rougher_raw', 'final_raw', 'rougher', 'final'], 
                          index=['Linear Regression', 'Ridge', 'Lasso', 'Tree', 'Forest'])
all_models
```

Модели будем рассматривать следующие: 
- линейная регрессия, 
- гребневая регрессия, 
- лассо-регрессия, 
- решающее дерево
- случайный лес. 

Чтобы получать хорошие значения метрики sMAPE, будем перебирать гиперпараметры для каждой модели жадным поиском GridSearchCV. Количество фолдов, на которых будет считаться кроссвалидационная оценка, возьмём 5 шт. А чтобы жадный поиск искал наименьшее значение нашей метрики, будем передавать параметру scoring название нашей метрики sMAPE, которую мы заранее написали.

Чтобы не писать один и тот же код 20 раз (по 5 моделей для необработанных и обработанных данных для каждой задачи), напишем функцию, которая по переданным параметрам аргументам понимает - для какой модели ей нужно искать гиперпараметры и для какой задачи (черновой концентрат или финальный).

```python
def search_params_and_metric(ml_model, params, X_raw, X, y, is_rough=False, score=sMAPE):
    
    # ml_model - модель, для которой будут искаться оптимальные гиперпараметры.
    # params - словарь гиперпараметров
    # X_raw - выборка всех признаков 
    # X - выборка некоррелирующих признаков
    # y - целевое значение
    
    # is_rough - параметр, означающий, что решается задача чернового концентрата. 
    # Если считаем черновой концентрат, is_rough установить True, eсли финальный - False
    start_time = time.time()
    
    # функция определяет, какую модель обучать
    if ml_model == 'Linear Regression':
        model = LinearRegression()
    elif ml_model == 'Ridge':
        model = Ridge(random_state=1212)
    elif ml_model == 'Lasso':
        model = Lasso(random_state=1212, max_iter=10**5)
    elif ml_model == 'Tree':
        model = DecisionTreeRegressor(random_state=1212, splitter='random')
    elif ml_model == 'Forest':
        model = RandomForestRegressor(random_state=1212, n_estimators=10, max_features='sqrt')
    else:
        return "Enter a correct ml_model's argument!"
    
    scaler = StandardScaler()
    pipe = Pipeline(steps=[("scaler", scaler), ("model", model)])
    search = GridSearchCV(estimator=pipe, param_grid=params, 
                                     cv=5, scoring=score)
    
    # служебные фразы для вывода лучших гиперпараметров
    phrase = f'''Лучшие гиперпараметры {ml_model} для предсказания финальной концентрации по данным'''
    if is_rough:
        phrase = f'''Лучшие гиперпараметры {ml_model} для предсказания черновой концентрации по данным'''

    search.fit(X_raw, y)
    print(phrase, 'со всеми исходными признаками:\n', search.best_params_)
    best_raw_score = -search.best_score_
    best_raw_model = search.best_estimator_
    print()
    
    search.fit(X, y)
    print(phrase, 'без коррелирующих признаков:\n', search.best_params_, end='\n\n')
    best_score = -search.best_score_
    best_model = search.best_estimator_
    
    print(f"--- {((time.time() - start_time)) / 60:.2f} --- minutes")
    
    return best_raw_model, best_model, best_raw_score, best_score
```
